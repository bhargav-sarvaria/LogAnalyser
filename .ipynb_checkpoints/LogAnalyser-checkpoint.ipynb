```python
import re
import pandas as pd
import time
```

```python
df = pd.DataFrame(columns=['time', 'proc_name','pid','operation', 'data', 'resource', 'object_id', 'direction'])
with open("filename.txt") as f:
    line_gen = zip(*[f]*3)
    for lines in line_gen:
        line_0 = lines[0]
        line_1 = lines[1].replace('(deleted)/ ', '')
        line_2 = lines[2]
        
        proc_name = lines[0].split(' ')[-2]
        event_time = lines[0].split(' ')[1]
        pid = line_1.split(' ')[0].replace('(', '').replace(')', '')
        operation = line_1.split(' ')[2]
        resource = line_1.split(' ')[4]
        
        direction = 1 if line_1.split(' ')[1] == '>' else 0
        
        object_id = 'na'
        
        if '<f>' in resource:
            object_id = re.search('<f>(.*)\)', resource).group(1)
            
        elif '<4u>' in resource:
            object_id = re.search('<4u>(.*)\)', resource).group(1)
            
        elif direction == 1:
            continue
        
        df.loc[len(df.index)] = [event_time, proc_name, pid, operation, line_1, resource, object_id, direction] 
        
temp_df = pd.DataFrame(columns=['time', 'proc_name','pid','operation', 'data', 'resource', 'object_id', 'direction'])
final_df = pd.DataFrame(columns=['start_time', 'end_time','subject','operation','object_id'])

for index, row in df.iterrows():
    try:
        if row['direction'] == 1:
            temp_df.loc[len(temp_df.index)] = row
        else:
            y = row
            x = temp_df.loc[(temp_df['proc_name'] == row['proc_name']) & (temp_df['pid']  == row['pid']) & (temp_df['operation']  == row['operation'])]
            ind = x.index
            x = x.iloc[0]
            subject = x['proc_name'] + '-' + x['pid']
            final_df.loc[len(final_df.index)] = [x['time'], y['time'], subject, x['operation'], x['object_id']] 
            temp_df.drop(ind , inplace=True)
    except Exception as e:
        continue
df = final_df
df.head()
```

```python
import pydot
def generateGraph(td, name):
    graph = pydot.Dot("my_graph", graph_type='digraph', bgcolor="white")
    subjects = td['subject'].unique()
    objects = td['object_id'].unique()
    
    for subject in subjects:
        graph.add_node(pydot.Node(str(subject), shape="circle"))

    for obj in objects:
        graph.add_node(pydot.Node(obj, shape="circle"))

    for index, row in td.iterrows():
        if row['operation'] in ['fcntl', 'write', 'sendmsg']:
            directed_edge = pydot.Edge(src=row['subject'], dst=row['object_id'], label = row['operation'])                                                                                
            graph.add_edge(directed_edge) 
        if row['operation'] in ['read', 'recvmsg']:
            directed_edge = pydot.Edge(src=row['object_id'], dst=row['subject'], label = row['operation'])                                                                                
            graph.add_edge(directed_edge) 

    graph.write_svg(name)
    
```

```python
generateGraph(df, "output.svg")
```

```python
def orderSubjects(td):
    tdf = pd.DataFrame(columns=['start_time', 'end_time','subject','operation', 'object_id'])

    for index, row in td.iterrows():
        if row['operation'] in ['read', 'recvmsg']:            
            tt = row['subject']
            row['subject'] = row['object_id']
            row['object_id'] = tt
            tdf.loc[len(tdf.index)] = row
        else:
            tdf.loc[len(tdf.index)] = row
    return tdf

def getPrevious(d, event):
    t1 = (d['object_id'] == event['subject'])&(d['start_time'] < event['end_time'])&(d['visited'] == False)
    d.loc[t1, 'visited'] = True
    filtered_df = d[t1]
    if filtered_df.shape[0] > 0:
        for index, row in filtered_df.iterrows():
            filtered_df = pd.concat([filtered_df, getPrevious(d, {'subject': row['subject'], 'end_time': row['end_time']})])
    return filtered_df.drop_duplicates()
```

```python
def getTrace(df, event):
    df = orderSubjects(df)
    df['visited'] = False
    backtrace_df = getPrevious(df, event).reset_index(drop=True)
    df = df.drop(['visited'], axis=1)
    backtrace_df = orderSubjects(backtrace_df)
    backtrace_df.to_csv('backtrace.csv')
    generateGraph(backtrace_df, str(int(time.time())) + '.svg')
    
# event = {'subject': 'spice-vdagentd-1307', 'end_time': '1668482992.516844751'}
# getTrace(df, event)
# df = orderSubjects(df)

event = {'subject': 'systemd-resolve-696', 'end_time': '1668482988.025725428'}
getTrace(df, event)
df = orderSubjects(df)


```
